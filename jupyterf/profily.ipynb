```python
%matplotlib inline
import pviz as pz
import random
no2 = pz.Particle("data/NO2_2003_2017.csv", "NO2")
o3 = pz.Particle("data/O3_2003_2017.csv", "O3")
pm10 = pz.Particle("data/PM10_2003_2017.csv", "pm10")
pm25 = pz.Particle("data/pm25_2003_2017.csv", "pm25")
```

```python
station = random.choice(no2.getStations())
no2.getCollectedByStation(station)
```

```python
no2.getCollectedYearOverall()
```

```python
pz.visualizeTimeProfile("hours",[no2],random.choice(no2.getStations()))
```

```python
pz.visualizeTimeProfile("months", [no2], random.choice(no2.getStations()))
```

```python
pz.visualizeTimeProfile("hours",[no2,o3,pm10,pm25],random.choice(list(no2.commonStation([o3,pm10,pm25]))))
```

```python
pz.visualizeTimeProfile("months",[no2,o3,pm10,pm25],random.choice(list(no2.commonStation([o3,pm10,pm25]))))
```
