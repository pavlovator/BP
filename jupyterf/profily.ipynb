```python
%matplotlib inline
import pviz as pz
import random
no2 = pz.Pollutant("data/NO2_2003_2017.csv", "NO2")
o3 = pz.Pollutant("data/O3_2003_2017.csv", "O3")
pm10 = pz.Pollutant("data/PM10_2003_2017.csv", "pm10")
pm25 = pz.Pollutant("data/pm25_2003_2017.csv", "pm25")
```

```python
station = random.choice(no2.getStations())
no2.getCoverageByStation(station)
```

```python
no2.getCoverageOverall()
```

```python
pz.visualizeTimeProfile("hours",[no2],random.choice(no2.getStations()))
```
